```python
# Celda 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')

# Celda 2: Cargar datos
orders = pd.read_csv('../data/raw/olist_orders_dataset.csv', parse_dates=['order_purchase_timestamp'])
orders['year_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
orders_by_month = orders['year_month'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
orders_by_month.plot(kind='bar')
plt.title('Órdenes por Mes', fontsize=16)
plt.xlabel('Mes')
plt.ylabel('Número de Órdenes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Celda 4: Órdenes por estado
customers = pd.read_csv('../data/raw/olist_customers_dataset.csv')
orders_customers = orders.merge(customers, on='customer_id')
state_orders = orders_customers['customer_state'].value_counts().head(10)

plt.figure(figsize=(10, 6))
state_orders.plot(kind='barh')
plt.title('Top 10 Estados por Órdenes', fontsize=16)
plt.xlabel('Número de Órdenes')
plt.tight_layout()
plt.show()
```
